## Zadanie 1

Zastosuj aproksymację wielomianem stopnia 1,2,3,...,20 do funkcji y = |x| na przedziale [−1, 1].

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return abs(x)

x = np.linspace(-1, 1, 101)
y = f(x)

for n in range(1, 21):
    p = np.polyfit(x, y, n)
    y_approx = np.polyval(p, x)
    error = np.mean(np.abs(y - y_approx))
    plt.plot(x, y_approx, label=f"Degree {n}, Error {error:.5f}")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Zadanie 4

Zastosuj aproksymację wielomianem sklejanym z rosnącą do 20 liczbą punktów kratowych do funkcji $y = \frac{1}{1+25x^2}$ na przedziale [−1, 1].

```python
import numpy as np
import scipy.interpolate as interp
import matplotlib.pyplot as plt

def f(x):
    return 1/(1+25*x**2)

x_approx = np.linspace(-1, 1, 20)
y = f(x_approx)

tck = interp.splrep(x_approx, y, s=0)
y_approx = interp.splev(x, tck, der=0)

plt.plot(x, y_approx, label="y = 1/(1+25x^2)")
plt.plot(x_approx, y, 'ro', label="Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Zadanie 5

Zastosuj aproksymację wielomianem szóstego stopnia dla funkcji $y = 40 + 10x + 5x2 + 3x3 + 2x4 +x5 +x6$, a właściwie dla jej wartości w punktach kratowych $a = 1 < 2 < 3 < 4 < 5 < 6 < 7 = b$. Czy zwiększenie liczby punktów kratowych na tym samym przedziale poprawia wynik aproksymacji?

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 40 + 10*x + 5*x**2 + 3*x**3 + 2*x**4 + x**5 + x**6

a = 1
b = 7
n = 20
x = np.linspace(a, b, n)
y = f(x)

p = np.polyfit(x, y, 6)
y_approx = np.polyval(p, x)

plt.plot(x, y, 'bo', label="f(x)")
plt.plot(x, y_approx, 'r-', label="Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
